```python
# Creating DataFrame
#df = pd.DataFrame(data)

# Custom function to classify term reasons
def classify_termreason(termreason):
    if termreason in [
        'more money', 'medical issues', 'maternity leave - did not return', 
        'return to school', 'relocation out of area', 'retiring', 
        'career change', 'unhappy', 'Another position'
    ]:
        return 'Voluntary'
    elif termreason in [
        'attendance', 'performance', 'gross misconduct', 
        'no-call, no-show', 'Learned that he is a gangster', 
        'Fatal attraction', 'military', 'hours'
    ]:
        return 'Involuntary'
    else:
        return 'Still Employed'

# Applying the function to the DataFrame
df['Employment_Status'] = df['TermReason'].apply(classify_termreason)

print(df)
```

```python
# Sort departments by median salary
mean_salary = df.groupby('Department')['Salary'].mean().sort_values(ascending=False)
sorted_departments = mean_salary.index

# Create a box plot with data labels
plt.figure(figsize=(12, 8))
sns.boxplot(x='Department', y='Salary', data=df, order=sorted_departments, palette='viridis')
plt.title('Department-wise Average Salary Distribution')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.xticks(rotation=45)

# Add average salary labels
for department in sorted_departments:
    avg_salary = mean_salary[department]
    plt.text(department, avg_salary, f'{avg_salary:.2f}', ha='center', va='center', fontsize=10, color='black', weight='bold')
plt.show()
```

```python
# Insights:

# The image is a box plot showing the department-wise average salary distribution. Each box plot displays the distribution of salaries within each department, including the median, quartiles, and outliers. These insights highlight the significant differences in average salaries across departments, with the Executive Office having the highest average salary and Production having the lowest.
```
