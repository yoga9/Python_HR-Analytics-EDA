```python
# Creating DataFrame
#df = pd.DataFrame(data)

# Custom function to classify term reasons
def classify_termreason(termreason):
    if termreason in [
        'more money', 'medical issues', 'maternity leave - did not return', 
        'return to school', 'relocation out of area', 'retiring', 
        'career change', 'unhappy', 'Another position'
    ]:
        return 'Voluntary'
    elif termreason in [
        'attendance', 'performance', 'gross misconduct', 
        'no-call, no-show', 'Learned that he is a gangster', 
        'Fatal attraction', 'military', 'hours'
    ]:
        return 'Involuntary'
    else:
        return 'Still Employed'

# Applying the function to the DataFrame
df['Employment_Status'] = df['TermReason'].apply(classify_termreason)

print(df)
```

```python
# Group by Performancescore and calculate the count for each score
grouped_df = df.groupby('PerformanceScore').size().sort_values(ascending=False).reset_index(name='count')

# Calculate the total count
total_count = df.shape[0]

# Calculate the percentage for each performance score
grouped_df['percentage'] = (grouped_df['count'] / total_count) * 100

print(grouped_df)
```

```python
# Create a bar plot
sns.barplot(x='percentage', y='PerformanceScore', data=grouped_df, palette='viridis')

# Add percentage labels
for index, row in grouped_df.iterrows():
    plt.text(row['percentage'], index, f"{row['percentage']:.2f}%", color='black', ha="center")



plt.title('Performance Scores Distribution')
plt.xlabel('Percentage')
plt.ylabel('Performance Score')
plt.show()
```

```python
# Insights:

# Performance Scores Distribution : Majority of employees are rated as "Fully Meets" or "Exceeds" expectations.
```
